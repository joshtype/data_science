```python
# TASK 0: import libraries for reading/manipulating/visualizing csv & data
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# TASK 1a: READ RAW CSV, IMPORT TO DF, PRINT 1st 10 ROWS
  # read raw csv
raw_csv = pd.read_csv("/content/drive/MyDrive/Data1501/Datasets/stroke.csv")
  # import csv to dataframe
df = pd.DataFrame(raw_csv)
  # output 1st 10 rows using iloc[] indexing
df.iloc[:10]
```

```python
# TASK 1b: FETCH 'avg_glucose_level' COL, PRINT 1st 10 ROWS
  # store avg_glucose_level col using iloc[] indexing
glucose_lvl = df.iloc[:, 8:9]
  # output first 10 rows of avg_glucose_level using head()
glucose_lvl.head(10)
```

```python
# TASK 2: GENERATE BASIC STATS FOR 'avg_glucose_level' COL
  # generate stats by invoking pandas' describe() method
glucose_lvl.describe()
```

```python
# TASK 3: GENERATE HISTOGRAM FROM COL DATA
  # invoke matplotlib.pyplot's hist() method, passing in col data
plt.hist(glucose_lvl, 25)
plt.title("Histogram of Average Glucose Levels")  # set title
plt.xlabel("Level of Glucose")                    # set x axis label
plt.ylabel("Number of People")                    # set y axis label
plt.xticks(range(50, 300, 25))                    # add xticks for clarity
plt.yticks(range(10, 180, 10))                    # add yticks for clarity
plt.show()                                        # show w/out code
  # comment with what the histogram says about the data
# The histogram shows that the data is skewed to the left. Meaning that
# levels trend toward lower values, ie, most people had low glucose levels.
# There is a distinct drop in values following the median value, leading to a
# noticeable outlier grouping of higher values, falling to the right side.
# This 'valley' then rises to a low peak at just above 30 people.
# The peak then trails to the extreme right at very large glucose levels.
# It suggests that most people stay within a lower range. However, for those
# whose glucose level does surpass the median, they are likely to do so
# significantly, peaking at levels between 175 to slightly above 200.
```

```python
# TASK 4a: FETCH LAST COL 'stroke' THEN COMMENT ON NUM STROKES
  # fetch stroke col
stroke_col = df['stroke']
  # output first 10 rows
stroke_col.head(10)
```

```python
# TASK 4b: GET VALUES FROM COL
  # can invoke sum() to add up total number of had-strokes (1s)
num_stroke = stroke_col.sum()
  # subtract sum from total num rows to get number of no-strokes (0s)
no_stroke = (len(stroke_col.index) - num_stroke)
  # print the values
print(f"{num_stroke} people repported having had a stroke.")
print(f"{no_stroke} people reported not having had a stroke.")
  # comment answer: 'How many ppl had stroke?' & 'How many didn't?'
# 137 people reported having had a stroke, while 999 people reported not having
# had a stroke.
```

```python
  # SUMMARY OF FINDINGS:
# The stroke data implies a correlation with the data shown in the histogram.
# The majority of people had lower glucose levels and the majority of people
# have never had a stroke. A significant but fractional number of people,
# had higher glucose levels and close to the same percentage of people
# reported having had a stroke. Digging deeper would most likely reveal that
# outliers of this correlation exist (lower level with stroke, higher level
# with no stroke). It would be interesting to compare the outliers with other
# reported data such as whether they smoke, age, etc. Regardless, the trend of
# the data makes it clear that higher glucose levels at least correlates with
# with increased risk of stroke.
```
