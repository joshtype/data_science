```python
# import matplotlib & pandas for reading/manip/visualizing csv & data
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# TASK #0: READ RAW CSV, LOAD RAW DATA TO DATAFRAME, OUTPUT DATAFRAME
  # 0: store raw data from csv dataset, indexed by year
raw = pd.read_csv("/content/drive/MyDrive/Data1501/Datasets/Headcounts.csv",
                  index_col = "Year")
  # 1: store raw data in dataframe
df = pd.DataFrame(raw)
  # 2: output dataframe
df
```

```python
# TASK #1: BAR CHART HEADCOUNTS IN 2012 ACROSS ALL DEPTS:

  # 0) extract necessary data from dataframe
dept = list(df.iloc[::])  # store list of ea department data
nums = df.loc[2012]       # store headcount numbers for 2012

  # 1) invoke bar() method passing dept for x, nums for y
plt.bar(dept, nums)

  # 3) set graph title, set labels for x axis, y axis, add legend
plt.title("Headcounts in Each Dept in 2012")  # graph title
plt.xlabel("Department")                      # x axis label
plt.ylabel("Headcount")                       # y axis label
plt.legend(["Staff Headcount"])               # add legend with title

  # 4) output finished graph without code as header
plt.show()  # invoke matplotlib's show()
```

```python
# TASK #2a: LINE CHART HEADCOUNTS ACROSS ALL DEPTS OVER ALL YEARS:

  # 0) invoke plot.line() to plot data by def index
df.plot.line()

  # 1) set graph title, x/y labels, legend added by default
plt.title("Changes in Total Headcounts Across all Departments (2010 - 2020):")
plt.xlabel("Year")                              # x axis label
plt.ylabel("Total Headcount")                   # y axis label

  # 2) output finished graph without code as header
plt.show()  # invoke matplotlib's show()
```

```python
# TASK #2b: LINE CHART HEADCOUNTS ACROSS ALL DEPTS OVER ALL YEARS:

  # 0) extract total headcounts from each dept per year, store in list
total_hc = list(df.sum(axis = 1))  # get sum() of ea row (1 = y axis)
years = list(df.index)

  # 1) load total_hc & years into a new dataframe
df_counts_yrs = pd.DataFrame({'Headcounts': total_hc}, index = years)

  # 2) invoke matplotlib's plot.line() with new dataframe
df_counts_yrs.plot.line()

  # 3) x ticks to show for all years/y ticks for range of all headcounts
plt.xticks(years)
plt.yticks(range(min(total_hc), max(total_hc)+1))

  # 3) set graph title, x/y labels, legend
plt.title("Changes in Total Headcounts Across all Departments (2010 - 2020):")
plt.xlabel("Year")                              # x axis label
plt.ylabel("Total Headcount")                   # y axis label
plt.legend(["Total Headcounts"])                # show legend w/ title

  # 4) output finished graph without code header
plt.show()  # invoke matplotlib's show()
```

```python
# TASK 3: PIE CHART DISTRIBUTION OF HEADCOUNTS ACROSS ALL DEPTS IN 2012

  # 0) using data stored in nums, invoke plot.pie(), set legend to show
nums.plot.pie(legend = True, labeldistance = None)  # remove unnecessary labels

  # 1) set graph title, x/y labels
plt.title("2012 Distribution of Headcounts Across Departments:")
plt.ylabel(None)  # not helpful information as year is in title

  # 2) output graph without code header
plt.show()  # invoke matplotlib's show()
```
